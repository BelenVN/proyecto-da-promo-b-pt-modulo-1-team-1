## <span style="color:red"> Hangman Game Star Wars Version ##

- Rules of the "Hangman Game":

    - One player (in this case, the computer) chooses a secret word and draws a space for each letter of the word.

    - The other player tries to guess letters to complete the word.

    - If the guessing player guesses a letter correctly, it is revealed in its corresponding place.

    - If the guessing player guesses incorrectly, a part of the body is drawn on the gallows.

    - The goal of the guessing player is to guess the word before the complete drawing is made on the gallows.

## <span style="color:green"> Version 1: no graphical interface ## 


#### It works well, but completely randomly, it sometimes skips a letter and only checks it when the next one is entered.

```python
# Import the random function to select a random word from the list of words.
import random

# Create a list of Star Wars themed words.
word_list = ["leia", "skywalker", "ewok", "chewbacca", "amidala", "mandalorian", "hyperspace"]
# Select a random word from the list of words.
secret_word = random.choice(word_list)

letter_list = list(secret_word) # Convert the secret word into a list with each of its letters.
unique_letters = set(letter_list)   # Create a set with the unique letters (that only appear once) of each secret word.
#print(secret_word)

# Drawing of the gallows as a list of strings, each drawing between the triple quotes is a string
# and which we can call with its indexes within the list

hanged = ['''
      +---+
      |   |
          |
          |       
          |
          |
    =========''', '''
      +---+
      |   |
      0   |
          |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
      |   |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|   |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|\  |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|\  |
     /    |
          |
    =========''', '''
      +---+
      |   |
      O   |
     /|\  |
     / \  |
          |
    =========''']




# This initializes the hyphenation board for letters converts it to a list. 
letter_spaces = '_' * len(secret_word)
letter_spaces = list(letter_spaces)

print("Avoid being hanged by the dark side, you must. Young Padawan, powerful in you, the Force is.")    

# Print board initial 
print(" ".join(letter_spaces))


# WHILE LOOP

# Variables we need: 
attempts = 0 # Game allows 6 attempts until the dummy hangs itself
tried_letters = [] # List of letters being tried to inform the player
playing = True # Flag to alert the while loop when the game is over
valid_letters = "abcdefghijklmnopqrstuvwxyz" # User input must be a letter


while attempts < 6 and playing: # As long as the attempts are less than 6 and playing remains True:
      
      user_letter = input("A single letter, you must enter, young Padawan: ").lower() # Ask the user to enter a letter and convert it to lower case
      if len(user_letter) != 1:
            print("Only a single letter, you must enter.")
            continue
      
      if user_letter not in valid_letters: # To let the user know that he/she/them has to use letters and not other types of characters
            print("Valid character, it is not. A letter, you must use.")
            continue 
      if user_letter in tried_letters: # Check if the letter has already been tried 
            print("Already tried this letter, you have, young Padawan. Another, you must choose.")
            continue 


      tried_letters.append(user_letter) # Add the letter entered by the user to the list of tried letters


      # Start conditionals      

      # Check if the user's letter is in the secret word
      if user_letter in letter_list:

            # For each matching letter, locate its index and substitute all instances into the secret word 
            for index, letter in enumerate(letter_list): 
                  if user_letter == letter: # When the user's letter matches a letter in the word: 
                        letter_spaces[index] = user_letter # The hyphen in the index position is replaced by the user's letter 
                
            unique_letters.discard(user_letter) # This removes it from the set of unique letters, leaving one less letter to guess 
                    
            print(" ".join(letter_spaces)) # Print the board again, and the lettering is already in place 
             
            print(f'Letters registered by R2D2 are: {tried_letters}') # We tell the user which letters he/she has already tried 
                        

            # Here it is checked whether all letters have been guessed 
            if len(unique_letters) == 0: # When the set of unique letters is empty all letters have been guessed 
                  playing = False # Change the condition that keeps the while loop running and exit the game 
                  print(f'Strong is the Force in you. Won, you have! {secret_word.upper()}, the secret word is.') # This prints the victory message 
      
      else: # If the letter is not in the word:
            attempts += 1 # Attempts was initialized to 0, now there is one less attempt left 
            print(f'In the word, the letter {user_letter} is not in') # Message is printed that the letter is not in the word 
            print(f'{6-attempts} attempts you have left, young padawan') # Message is printed how many attempts are left 
            # For the strings in the hangman drawing, print the one at the index that matches the number of attempts 
            # In the 1st attempt the index 1, which is the second drawing, that is to say it adds the head, and so on 
            print(hanged[attempts]) # The hanged man's dummy is printed in its corresponding state 
            print(f'Letters registered by R2D2 are: {tried_letters}')
            
            

# When the number of attempts reaches 6, the player has lost and is kicked out of the game 
if attempts == 6:
      print(f'The dark side, won it has. Train more, you must, young Padawan. The secret word, {secret_word.upper()} it was.') # It indicates what the secret word was 
```
